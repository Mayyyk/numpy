```python
import numpy as np
import matplotlib.pyplot as plt

# Prosty przykad - wzrost wykadniczy
x = np.linspace(0, 10, 100)
y = np.exp(x)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Zwyka skala
axes[0].plot(x, y, 'b-', linewidth=2)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y = e^x')
axes[0].set_title('ZWYKA SKALA\nWzrost wykadniczy wyglda... wykadniczo')
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(0, 22000)

# Skala logarytmiczna
axes[1].semilogy(x, y, 'r-', linewidth=2)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y = e^x (skala log)')
axes[1].set_title('SKALA LOGARYTMICZNA\nWzrost wykadniczy wyglda LINIOWO!')
axes[1].grid(True, which='both', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Wizualizacja jak dziaa o logarytmiczna
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# O liniowa
axes[0].plot([1, 2, 3, 4, 5], [0, 0, 0, 0, 0], 'bo', markersize=10)
axes[0].set_xlim(0, 6)
axes[0].set_ylim(-1, 1)
axes[0].set_title('O LINIOWA: rwne odlegoci = dodawanie')
axes[0].set_xlabel('1  2  3  4  5  (dodajemy +1 kady krok)')
axes[0].set_yticks([])
axes[0].grid(True, axis='x')

# O logarytmiczna
axes[1].semilogx([1, 10, 100, 1000, 10000], [0, 0, 0, 0, 0], 'ro', markersize=10)
axes[1].set_xlim(0.5, 20000)
axes[1].set_ylim(-1, 1)
axes[1].set_title('O LOGARYTMICZNA: rwne odlegoci = mnoenie')
axes[1].set_xlabel('1  10  100  1000  10000  (mnoymy 10 kady krok)')
axes[1].set_yticks([])
axes[1].grid(True, which='both', axis='x')

plt.tight_layout()
plt.show()

print("Na osi logarytmicznej:")
print(" Odlego od 1 do 10 = taka sama jak od 10 do 100")
print(" Odlego od 10 do 100 = taka sama jak od 100 do 1000")
print(" Kada rwna odlego = mnoenie przez t sam liczb")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 100, 1000)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Przykad 1: y = x (wzrost liniowy)
axes[0, 0].loglog(x, x, 'b-', linewidth=2)
axes[0, 0].set_title('y = x\n(wzrost LINIOWY)')
axes[0, 0].set_xlabel('x (log)')
axes[0, 0].set_ylabel('y (log)')
axes[0, 0].grid(True, which='both', alpha=0.3)
axes[0, 0].text(5, 50, 'Linia prosta pod 45\n= wzrost LINIOWY', 
                bbox=dict(boxstyle='round', facecolor='wheat'))

# Przykad 2: y = x (prawo potgowe)
axes[0, 1].loglog(x, x**2, 'g-', linewidth=2)
axes[0, 1].set_title('y = x\n(prawo POTGOWE)')
axes[0, 1].set_xlabel('x (log)')
axes[0, 1].set_ylabel('y (log)')
axes[0, 1].grid(True, which='both', alpha=0.3)
axes[0, 1].text(5, 1000, 'Linia prosta\n= prawo potgowe\nNachylenie = wykadnik', 
                bbox=dict(boxstyle='round', facecolor='lightgreen'))

# Przykad 3: y = e^x (wzrost wykadniczy)
x2 = np.linspace(0.1, 10, 100)
axes[0, 2].semilogy(x2, np.exp(x2), 'r-', linewidth=2)
axes[0, 2].set_title('y = e^x\n(wzrost WYKADNICZY)')
axes[0, 2].set_xlabel('x (liniowy)')
axes[0, 2].set_ylabel('y (log)')
axes[0, 2].grid(True, which='both', alpha=0.3)
axes[0, 2].text(2, 10000, 'Linia prosta\n= wzrost WYKADNICZY', 
                bbox=dict(boxstyle='round', facecolor='lightcoral'))

# Przykad 4: y = log(x)
axes[1, 0].semilogx(x, np.log10(x), 'm-', linewidth=2)
axes[1, 0].set_title('y = log(x)\n(wzrost LOGARYTMICZNY)')
axes[1, 0].set_xlabel('x (log)')
axes[1, 0].set_ylabel('y (liniowy)')
axes[1, 0].grid(True, which='both', alpha=0.3)
axes[1, 0].text(10, 1.5, 'Linia prosta\n= wzrost LOGARYTMICZNY', 
                bbox=dict(boxstyle='round', facecolor='plum'))

# Przykad 5: y = 1/x (odwrotna proporcjonalno)
axes[1, 1].loglog(x, 1/x, 'c-', linewidth=2)
axes[1, 1].set_title('y = 1/x\n(ODWROTNA proporcjonalno)')
axes[1, 1].set_xlabel('x (log)')
axes[1, 1].set_ylabel('y (log)')
axes[1, 1].grid(True, which='both', alpha=0.3)
axes[1, 1].text(5, 0.5, 'Linia prosta\nz nachyleniem -1', 
                bbox=dict(boxstyle='round', facecolor='lightcyan'))

# Przykad 6: y = staa (brak zalenoci)
axes[1, 2].loglog(x, 10*np.ones_like(x), 'orange', linewidth=2)
axes[1, 2].set_title('y = 10 (staa)\n(BRAK zalenoci)')
axes[1, 2].set_xlabel('x (log)')
axes[1, 2].set_ylabel('y (log)')
axes[1, 2].grid(True, which='both', alpha=0.3)
axes[1, 2].text(5, 15, 'Linia POZIOMA\n= brak zalenoci', 
                bbox=dict(boxstyle='round', facecolor='moccasin'))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Symulacja wzrostu liczby przypadkw (wzrost wykadniczy na pocztku)
dni = np.arange(1, 100)
przypadki = 10 * np.exp(0.15 * dni)  # Pocztkowo wykadniczy wzrost

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Zwyka skala - trudno zobaczy wczesny wzrost
axes[0].plot(dni, przypadki, 'r-', linewidth=2)
axes[0].set_xlabel('Dni od pierwszego przypadku')
axes[0].set_ylabel('Liczba przypadkw')
axes[0].set_title('ZWYKA SKALA\nWczesny wzrost niewidoczny!')
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Pierwsze 20 dni\nnie wida prawie nic!', 
                 xy=(10, przypadki[9]), xytext=(30, 1e8),
                 arrowprops=dict(arrowstyle='->', color='blue', lw=2))

# Skala logarytmiczna - wyranie wida wzrost wykadniczy
axes[1].semilogy(dni, przypadki, 'b-', linewidth=2)
axes[1].set_xlabel('Dni od pierwszego przypadku')
axes[1].set_ylabel('Liczba przypadkw (skala log)')
axes[1].set_title('SKALA LOGARYTMICZNA\nLinia prosta = wzrost wykadniczy!')
axes[1].grid(True, which='both', alpha=0.3)
axes[1].annotate('Prosta linia = stae tempo wzrostu\n(podwajanie co X dni)', 
                 xy=(50, przypadki[49]), xytext=(30, 1e4),
                 arrowprops=dict(arrowstyle='->', color='red', lw=2))

plt.tight_layout()
plt.show()

print("\nINTERPRETACJA:")
print(" Linia prosta na skali log = wzrost wykadniczy (kady dzie +15%)")
print(" Jeli linia zaczyna si zakrzywia w d = spowolnienie wzrostu")
print(" Jeli linia jest pozioma = liczba przypadkw staa")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Prawo Moore'a: liczba tranzystorw podwaja si co ~2 lata
lata = np.array([1971, 1978, 1985, 1993, 2000, 2008, 2017, 2023])
tranzystory = np.array([2300, 29000, 275000, 3100000, 42000000, 
                        731000000, 19200000000, 100000000000])
nazwy = ['4004', '8086', '80386', 'Pentium', 'Pentium 4', 
         'Core i7', 'Epyc', 'M2 Ultra']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Zwyka skala - tylko ostatnie procesory widoczne
axes[0].plot(lata, tranzystory, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Rok')
axes[0].set_ylabel('Liczba tranzystorw')
axes[0].set_title('ZWYKA SKALA - wczesne procesory niewidoczne')
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Wszystkie procesory\ndo 2000 wygldaj na 0!', 
                 xy=(1985, tranzystory[2]), xytext=(1990, 4e10),
                 arrowprops=dict(arrowstyle='->', color='red', lw=2))

# Skala logarytmiczna - wyrana LINIA PROSTA = prawo Moore'a dziaa!
axes[1].semilogy(lata, tranzystory, 'ro-', linewidth=2, markersize=8)
axes[1].set_xlabel('Rok')
axes[1].set_ylabel('Liczba tranzystorw (skala log)')
axes[1].set_title('SKALA LOG - Prawo Moore\'a jako prosta linia!')
axes[1].grid(True, which='both', alpha=0.3)

# Dodaj etykiety
for i, (rok, trans, nazwa) in enumerate(zip(lata, tranzystory, nazwy)):
    axes[1].annotate(nazwa, xy=(rok, trans), xytext=(5, 5), 
                    textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()

print("\nINTERPRETACJA PRAWA MOORE'A:")
print(" Linia prosta na skali log = podwajanie co ~2 lata (wzrost wykadniczy)")
print(" Moemy przewidzie przyszo - przeduenie linii")
print(" Jeli linia si wypaszcza = prawo Moore'a si koczy")
```
