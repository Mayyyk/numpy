```python
import numpy as np
import matplotlib.pyplot as plt

# Prosty przykład - wzrost wykładniczy
x = np.linspace(0, 10, 100)
y = np.exp(x)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Zwykła skala
axes[0].plot(x, y, 'b-', linewidth=2)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y = e^x')
axes[0].set_title('ZWYKŁA SKALA\nWzrost wykładniczy wygląda... wykładniczo')
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(0, 22000)

# Skala logarytmiczna
axes[1].semilogy(x, y, 'r-', linewidth=2)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y = e^x (skala log)')
axes[1].set_title('SKALA LOGARYTMICZNA\nWzrost wykładniczy wygląda LINIOWO!')
axes[1].grid(True, which='both', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Wizualizacja jak działa oś logarytmiczna
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Oś liniowa
axes[0].plot([1, 2, 3, 4, 5], [0, 0, 0, 0, 0], 'bo', markersize=10)
axes[0].set_xlim(0, 6)
axes[0].set_ylim(-1, 1)
axes[0].set_title('OŚ LINIOWA: równe odległości = dodawanie')
axes[0].set_xlabel('1 → 2 → 3 → 4 → 5  (dodajemy +1 każdy krok)')
axes[0].set_yticks([])
axes[0].grid(True, axis='x')

# Oś logarytmiczna
axes[1].semilogx([1, 10, 100, 1000, 10000], [0, 0, 0, 0, 0], 'ro', markersize=10)
axes[1].set_xlim(0.5, 20000)
axes[1].set_ylim(-1, 1)
axes[1].set_title('OŚ LOGARYTMICZNA: równe odległości = mnożenie')
axes[1].set_xlabel('1 → 10 → 100 → 1000 → 10000  (mnożymy ×10 każdy krok)')
axes[1].set_yticks([])
axes[1].grid(True, which='both', axis='x')

plt.tight_layout()
plt.show()

print("Na osi logarytmicznej:")
print("• Odległość od 1 do 10 = taka sama jak od 10 do 100")
print("• Odległość od 10 do 100 = taka sama jak od 100 do 1000")
print("• Każda równa odległość = mnożenie przez tę samą liczbę")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 100, 1000)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Przykład 1: y = x (wzrost liniowy)
axes[0, 0].loglog(x, x, 'b-', linewidth=2)
axes[0, 0].set_title('y = x\n(wzrost LINIOWY)')
axes[0, 0].set_xlabel('x (log)')
axes[0, 0].set_ylabel('y (log)')
axes[0, 0].grid(True, which='both', alpha=0.3)
axes[0, 0].text(5, 50, 'Linia prosta pod 45°\n= wzrost LINIOWY', 
                bbox=dict(boxstyle='round', facecolor='wheat'))

# Przykład 2: y = x² (prawo potęgowe)
axes[0, 1].loglog(x, x**2, 'g-', linewidth=2)
axes[0, 1].set_title('y = x²\n(prawo POTĘGOWE)')
axes[0, 1].set_xlabel('x (log)')
axes[0, 1].set_ylabel('y (log)')
axes[0, 1].grid(True, which='both', alpha=0.3)
axes[0, 1].text(5, 1000, 'Linia prosta\n= prawo potęgowe\nNachylenie = wykładnik', 
                bbox=dict(boxstyle='round', facecolor='lightgreen'))

# Przykład 3: y = e^x (wzrost wykładniczy)
x2 = np.linspace(0.1, 10, 100)
axes[0, 2].semilogy(x2, np.exp(x2), 'r-', linewidth=2)
axes[0, 2].set_title('y = e^x\n(wzrost WYKŁADNICZY)')
axes[0, 2].set_xlabel('x (liniowy)')
axes[0, 2].set_ylabel('y (log)')
axes[0, 2].grid(True, which='both', alpha=0.3)
axes[0, 2].text(2, 10000, 'Linia prosta\n= wzrost WYKŁADNICZY', 
                bbox=dict(boxstyle='round', facecolor='lightcoral'))

# Przykład 4: y = log(x)
axes[1, 0].semilogx(x, np.log10(x), 'm-', linewidth=2)
axes[1, 0].set_title('y = log(x)\n(wzrost LOGARYTMICZNY)')
axes[1, 0].set_xlabel('x (log)')
axes[1, 0].set_ylabel('y (liniowy)')
axes[1, 0].grid(True, which='both', alpha=0.3)
axes[1, 0].text(10, 1.5, 'Linia prosta\n= wzrost LOGARYTMICZNY', 
                bbox=dict(boxstyle='round', facecolor='plum'))

# Przykład 5: y = 1/x (odwrotna proporcjonalność)
axes[1, 1].loglog(x, 1/x, 'c-', linewidth=2)
axes[1, 1].set_title('y = 1/x\n(ODWROTNA proporcjonalność)')
axes[1, 1].set_xlabel('x (log)')
axes[1, 1].set_ylabel('y (log)')
axes[1, 1].grid(True, which='both', alpha=0.3)
axes[1, 1].text(5, 0.5, 'Linia prosta\nz nachyleniem -1', 
                bbox=dict(boxstyle='round', facecolor='lightcyan'))

# Przykład 6: y = stała (brak zależności)
axes[1, 2].loglog(x, 10*np.ones_like(x), 'orange', linewidth=2)
axes[1, 2].set_title('y = 10 (stała)\n(BRAK zależności)')
axes[1, 2].set_xlabel('x (log)')
axes[1, 2].set_ylabel('y (log)')
axes[1, 2].grid(True, which='both', alpha=0.3)
axes[1, 2].text(5, 15, 'Linia POZIOMA\n= brak zależności', 
                bbox=dict(boxstyle='round', facecolor='moccasin'))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Symulacja wzrostu liczby przypadków (wzrost wykładniczy na początku)
dni = np.arange(1, 100)
przypadki = 10 * np.exp(0.15 * dni)  # Początkowo wykładniczy wzrost

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Zwykła skala - trudno zobaczyć wczesny wzrost
axes[0].plot(dni, przypadki, 'r-', linewidth=2)
axes[0].set_xlabel('Dni od pierwszego przypadku')
axes[0].set_ylabel('Liczba przypadków')
axes[0].set_title('ZWYKŁA SKALA\nWczesny wzrost niewidoczny!')
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Pierwsze 20 dni\nnie widać prawie nic!', 
                 xy=(10, przypadki[9]), xytext=(30, 1e8),
                 arrowprops=dict(arrowstyle='->', color='blue', lw=2))

# Skala logarytmiczna - wyraźnie widać wzrost wykładniczy
axes[1].semilogy(dni, przypadki, 'b-', linewidth=2)
axes[1].set_xlabel('Dni od pierwszego przypadku')
axes[1].set_ylabel('Liczba przypadków (skala log)')
axes[1].set_title('SKALA LOGARYTMICZNA\nLinia prosta = wzrost wykładniczy!')
axes[1].grid(True, which='both', alpha=0.3)
axes[1].annotate('Prosta linia = stałe tempo wzrostu\n(podwajanie co X dni)', 
                 xy=(50, przypadki[49]), xytext=(30, 1e4),
                 arrowprops=dict(arrowstyle='->', color='red', lw=2))

plt.tight_layout()
plt.show()

print("\nINTERPRETACJA:")
print("• Linia prosta na skali log = wzrost wykładniczy (każdy dzień +15%)")
print("• Jeśli linia zaczyna się zakrzywiać w dół = spowolnienie wzrostu")
print("• Jeśli linia jest pozioma = liczba przypadków stała")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Prawo Moore'a: liczba tranzystorów podwaja się co ~2 lata
lata = np.array([1971, 1978, 1985, 1993, 2000, 2008, 2017, 2023])
tranzystory = np.array([2300, 29000, 275000, 3100000, 42000000, 
                        731000000, 19200000000, 100000000000])
nazwy = ['4004', '8086', '80386', 'Pentium', 'Pentium 4', 
         'Core i7', 'Epyc', 'M2 Ultra']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Zwykła skala - tylko ostatnie procesory widoczne
axes[0].plot(lata, tranzystory, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Rok')
axes[0].set_ylabel('Liczba tranzystorów')
axes[0].set_title('ZWYKŁA SKALA - wczesne procesory niewidoczne')
axes[0].grid(True, alpha=0.3)
axes[0].annotate('Wszystkie procesory\ndo 2000 wyglądają na 0!', 
                 xy=(1985, tranzystory[2]), xytext=(1990, 4e10),
                 arrowprops=dict(arrowstyle='->', color='red', lw=2))

# Skala logarytmiczna - wyraźna LINIA PROSTA = prawo Moore'a działa!
axes[1].semilogy(lata, tranzystory, 'ro-', linewidth=2, markersize=8)
axes[1].set_xlabel('Rok')
axes[1].set_ylabel('Liczba tranzystorów (skala log)')
axes[1].set_title('SKALA LOG - Prawo Moore\'a jako prosta linia!')
axes[1].grid(True, which='both', alpha=0.3)

# Dodaj etykiety
for i, (rok, trans, nazwa) in enumerate(zip(lata, tranzystory, nazwy)):
    axes[1].annotate(nazwa, xy=(rok, trans), xytext=(5, 5), 
                    textcoords='offset points', fontsize=8)

plt.tight_layout()
plt.show()

print("\nINTERPRETACJA PRAWA MOORE'A:")
print("• Linia prosta na skali log = podwajanie co ~2 lata (wzrost wykładniczy)")
print("• Możemy przewidzieć przyszłość - przedłużenie linii")
print("• Jeśli linia się wypłaszcza = prawo Moore'a się kończy")
```
